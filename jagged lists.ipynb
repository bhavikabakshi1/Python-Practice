```python
grid = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
total=sum(sum(row) for row in grid)
print("sum of all elements:", total)
count=sum(len(row) for row in grid)
average = total/ count

print("average of all elements:" , average)
min_val = grid[0][0]
max_val = grid[0][0]

for row in grid:
    for value in row:
        if value < min_val:
            min_val= value
        if value > max_val:
            max_val = value
print("minimum elements: " , min_val)
print("maximum elemets: " ,max_val)
```

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
value = int(input("Enter the value to search: "))
found = False
for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        if matrix[i][j] == value:
            print(f"Element {value} found at Row {i} and Column {j}")
            break
            if found :
                break
            found = True
if not found:
    print(f"Element {value} not found in the list.")
```

```python
matrix = [
    [1, 2, 3],
    [3 , 4 ,5],
    [7, 7 , 7]
]
x=5
positions = []

for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        if matrix[i][j] == x:
            positions.append(i,j)
            
if positions:
    print(f" element{x} found at positions: {positions})
else:
    print(f"element{x} not found ")
```
