```python

List2 = ["Garima", "GUrU", "RAMa", "HEY", "how", "are", "YOU"]

small_counts = {}
big_counts = {}

for word in List2:
    for char in word:
        if char.islower(): 
            small_counts[char] = small_counts.get(char, 0) + 1
        elif char.isupper():
            big_counts[char] = big_counts.get(char, 0) + 1

print("Lowercase letters:")
for char, count in small_counts.items():
    print(f"{char}:{count}", end=", ")
print("\n")

print("Uppercase letters:")
for char, count in big_counts.items():
    print(f"{char}:{count}", end=", ")
print("\n")

print("Total lowercase letters:", sum(small_counts.values()))
print("Total uppercase letters:", sum(big_counts.values()))
```

```python
def tower_of_hanoi(n, source, auxillary, target):
    if n ==1:
        print(f"move disk 1 from {source} to {target}")
        return 1
    moves= tower_of_hanoi(n-1, source, target, auxillary)
    print(f"move disk {n} from {source} to {target}")
    moves +=1
    moves+= tower_of_hanoi(n-1,auxillary,source,target)
    return moves

n=int(input("enter the number of disk:"))
print(f"tower of hanoi solution for {n} disks(source= A, auxillary= C, target= B):\n")
total_moves= tower_of_hanoi(n,'A','C','B')
print(f"\ntotal moves required:{total_moves}")
```

```python
def gcd(a,b):
    while b!=0:
        a,b = b, a%b
    return a
num1=int(input("enter first number:"))
num2= int(input("enter the second number:"))
print("gcd of",num1,"and",num2 ,"is: ",gcd(num1,num2))
```

```python
def gcd_three(a,b,c):
    while b!=0:
        a,b = b, a%b
    while c!=0:
        a,c=c,a%c
    return a
num1=int(input("enter first number:"))
num2= int(input("enter the second number:"))
num3= int(input("enetr the third number:"))
print("gcd of",num1,",",num2,"and",num3,"is:",gcd_three(num1,num2,num3))
```

```python
#matrices
L1 =[[1,2,3,4],
     [3,4,5,6]]
L2=[[1,2,3,4],
    [3,4,5,6]]
#matrix mulitiplication::
result=[[0 for _ in range(len(L2[0]))] for _ in range(len(L1))]

for i in range(len(L1)):
    for j in range(len(L2[0])):
        for k in range(len(L2)):
            result[i][j] += L1[i][k] * L2[k][j]

for row in result:
    print(row)
```

```python
# find max and min length from  a 2d jagged list
def find_max(list1):
  return max(max(row) for row in list1)

def find_min(list1):
  return min(min(row) for row in list1)
list1 = [
      [21,12],
      [43,9,16],[27,38,5,13]
      ]
maximum_number= find_max(list1)
minimum_number= find_min(list1)
print(" maximum number: " , maximum_number)
print(" minimum number:", minimum_number)
```
