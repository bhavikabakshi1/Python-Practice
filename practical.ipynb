```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
num = int(input("Enter a positive integer: "))
sum_of_factorials = 0
for digit in str(num):
    sum_of_factorials += factorial(int(digit))
if sum_of_factorials == num:
    print(" it is is a Strong number.")
else:
    print("it is not a Strong number.")
```

```python
n = int(input("Enter a positive integer: "))
a, b = 0, 1
fibonacci_sequence = []
sum_fib = 0
while a <= n:
    fibonacci_sequence.append(a)
    sum_fib += a
    a, b = b, a + b
print("Fibonacci sequence up to", n, "is:", fibonacci_sequence)
print("Sum of Fibonacci numbers up to", n, "=", sum_fib)
```

```python
#13
filename = input("Enter the file name: ")
try:
    f = open(filename, "r")
    data = f.read()
    f.close()
    data = data.lower()
    punct = ",.!?;:'\"()[]{}<>-"
    for p in punct:
        data = data.replace(p, "")
    words = data.split()
    freq = {}
    for w in words:
        if w in freq:
            freq[w] += 1
        else:
            freq[w] = 1
    max_word = None
    max_count = 0
    for key in freq:
        if freq[key] > max_count:
            max_count = freq[key]
            max_word = key
    print("\nWord Frequency Dictionary:")
    for k in freq:
        print(k, ":", freq[k])

    print("\nMost Frequent Word:")
    print(max_word, "appeared", max_count, "times")

except FileNotFoundError:
    print("File not found. Please check the file name.")
```

```python
#15
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks
    def display_details(self):
        print("Student Name:", self.name)
        print("Roll Number:", self.roll_number)
        print("Marks:", self.marks)
    def check_result(self):
        if self.marks >= 40:
            print("Result: Passed")
        else:
            print("Result: Failed")
name = input("Enter student name: ")
roll_number = int(input("Enter roll number: "))
marks = int(input("Enter marks out of 100: "))
if marks < 0 or marks > 100:
    print("Marks should be between 0 and 100")
else:
    student1 = Student(name, roll_number, marks)
    student1.display_details()
    student1.check_result()
```

```python
#16
def add(a, b):
    return a + b
def sub(a, b):
    return a - b
def mul(a, b):
    return a * b
def div(a, b):
    return a / b
print("Simple Calculator")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
try:
    choice = int(input("Enter your choice (1-4): "))
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    if choice == 1:
        print("Result =", add(num1, num2))
    elif choice == 2:
        print("Result =", sub(num1, num2))
    elif choice == 3:
        print("Result =", mul(num1, num2))
    elif choice == 4:
        try:
            print("Result =", div(num1, num2))
        except ZeroDivisionError:
            print("Error: Division by zero is not allowed")
    else:
        print("Invalid Choice")
except ValueError:
    print("Invalid Input! Please enter numbers only.")
except Exception as e:
    print("Some error occurred:", e)
```
