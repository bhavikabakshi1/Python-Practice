```python
def anagram_check(word1,word2):
    word1=word1.lower()
    word2=word2.lower()
    sorted1=sorted(word1)
    sorted2=sorted(word2)
    return sorted1==sorted2
print(anagram_check("listen","silent"))
print(anagram_check("heart","earth"))
```

```python
def message_flipper(message):
    flipped= ""
    for ch in message:
        if ch.islower():
            flipped += ch.upper()
        elif ch.isupper():
            flipped += ch.lower()
        else:
            flipped+=ch
    return flipped
message=input().strip()
print(message_flipper(message))
    
        
```

```python
def word_replacment(sentence,old_word, new_word):
    upd_sent=sentence.replace(old_word, new_word)
    return upd_sent
print(word_replacment("hey how are you","hey","hi"))
```

```python
def avg_ratings(ratings):
    if not ratings:
        return 0
    else:
        return sum(ratings)/len(ratings)
print(avg_ratings([5, 4, 3, 4, 5]))

```

```python
def highest_marks(marks):
    if not marks:
        return -0
    highest=marks[0]
    for num in marks:
        if num is highest:
            highest= num
            return highest
print (highest_marks([10, 20, 30]))
```
